```python

# Start Driver, get site, create link list

uni_links = [] # list for collection of of sites containing information
driver = webdriver.Chrome(ChromeDriverManager().install()) # check for latest Webdriver

driver.get("https://studieren.de/hochschulliste.0.html") # driver navigates to site

#page = driver.page_source
links = driver.find_elements_by_class_name("item-main-link") # find links to sites on one page

for l in links:   #collect links from one page
    link = l.get_attribute("href")
    uni_links.append(link)

for site_nr in range(2,24): #iterate through sites with links
    link = "https://studieren.de/hochschulliste.t-0.s-" + str(site_nr) + ".html" # build link
    driver.get(link) # load site
    links = driver.find_elements_by_class_name("item-main-link") #collect links
    for l in links:
        link = l.get_attribute("href")
        uni_links.append(link)

print(uni_links)

```

```python
# check if list is working and exclude wrong links
new = [x for x in uni_links if x != "https://secure.studieren.de/goto/3747/ServiceChannelsUniversities/"] 
print(len(new))
#some unis have double entries --> solve with address comparison later (found at ovgu ironically)

```
