```python
# set up dataframe with all variables I want to scrape
import pandas as pd
data_cols = ['id_uni', 'name_uni', 'city', 'plz', 'county', 'address',  'traeger', 'nr_students', 'found_year', 'long', 'lat', 'ap_auf', 'ap_ing', 'ap_num', 'ap_mug', 'ap_gus', 'ap_ruw', 'ap_sloth', 'ap_suk', 'ap_kgm', 'ap_lsg', 'studies']
df = pd.DataFrame(columns=data_cols)
df

```

```python
# loop that iterates through the cleaned list of links and collects all the info into a dataframe

driver = webdriver.Chrome(ChromeDriverManager().install()) # start driver

for u in links_cleaned:

    driver.get(u) # get data site

    # get infos from site

    # id 
    id_uni = u.split('.')[-2] # take id out of link from studieren.de
    #print(id_uni)

    # name uni
    name_uni = driver.find_element_by_tag_name("h1").text   # finds <h1> html tak | .text to extract text between html tags
    #print(name_uni)

    # name uni
    city = driver.find_element_by_class_name("headlines-additional-info").text  # finds html tag with specified class name | .text to extract text between html tags
    #print(city)

    # studies focus
    study_focs = []
    element = driver.find_elements_by_xpath("//span[@class='listing-sciences-item-cell value']") # finds all elements that are <span> with specified class name
    
    # extract text from all elements found by xpath. Since the order of the info is the same on every site, 
    # the indices of the list is used to store the info in the right variable
    for v in element: 
        study_focs.append(v.text)    
    #print(study_focs)
    ap_auf = study_focs[0]
    ap_ing = study_focs[1]
    ap_num = study_focs[2]
    ap_mug = study_focs[3]
    ap_gus = study_focs[4]
    ap_ruw = study_focs[5]
    ap_sloth = study_focs[6]
    ap_suk = study_focs[7]
    ap_kgm = study_focs[8]
    ap_lsg = study_focs[9]

    # long lat
    # Whole page code is collected. Info is located in text. Split is used to cut off unneeded text
    page = driver.page_source
    long_lat = page.split("new gm.LatLng(")[1]
    long_lat = long_lat.split(")")[0]
    long = long_lat.split(',')[0]
    lat = long_lat.split(',')[1]
    #print(long_lat)


    # Uni details
    # uni attribute
    # Info about uni differs for some sites. Variables that need to be collected are defined. 
    # Info from site is collected into dictionary. If input for variable is found in the dictionary, it is stored in the variable.
    uni_attrs = []
    uni_vals = []
    uni_attr = driver.find_elements_by_xpath("//div[@class='simple-inline-table-cell label']")
    for u in uni_attr:
        uni_attrs.append(u.text)
    # Value values: 
    uni_val = driver.find_elements_by_xpath("//div[@class='simple-inline-table-cell value']")
    for u in uni_val:
        uni_vals.append(u.text)
    #print(uni_attrs)
    #print(uni_vals)

    uni_combined = dict(zip(uni_attrs, uni_vals))
    #print(uni_combined)
    
    # reset variables 
    plz = "NaN"
    city = "NaN"
    county = "NaN"
    traeger = "NaN"
    nr_students = "NaN"
    found_year = "NaN"

    # Uni Attributes
    for a in uni_attrs:
        if a == "Stadt":
            plz = uni_combined["Stadt"].split(" ")[0]
            city = uni_combined["Stadt"].split(" ")[1]    
        elif a == "Bundesland":
            county = uni_combined["Bundesland"]
        elif a == "Sekretariat":
            address = uni_combined["Sekretariat"]
        elif a == "Trgerschaft":
            traeger = uni_combined["Trgerschaft"]
        elif a == "Studierende":
            nr_students = uni_combined["Studierende"]
        elif a == "Grndungsjahr":
            found_year = uni_combined["Grndungsjahr"]
        


    # Available studies
    # name of studies
    # All studies available at the uni are located and stored in a list.
    study_names = []
    study_types = []
    study_list = []
    study_name = driver.find_elements_by_xpath("//div[@class='primary-headline']")
    for u in study_name:
        study_names.append(u.text)
    # Value values: 
    study_type = driver.find_elements_by_xpath("//span[@class='text-switch-long']")
    for u in study_type:
        study_types.append(u.text)
    print(study_names)
    print(study_types)

    for a,b in zip(study_names,study_types):
        study_list.append(a + ", " + b)
    studies = study_list

    # insert Variables as column into dataframe. After every iteration a new column with all the newly defined variables is stored in the dataframe.
    df.loc[len(df.index)] = [id_uni, name_uni, city, plz, county, address, traeger, nr_students, found_year, long, lat, ap_auf, ap_ing, ap_num, ap_mug, ap_gus, ap_ruw, ap_sloth, ap_suk, ap_kgm, ap_lsg, studies]
    
#df
```
